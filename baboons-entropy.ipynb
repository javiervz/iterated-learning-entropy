```python
def neighbors(matrix, coordinates, distance=1):
    dimensions = len(coordinates)
    neigh = []
    app = neigh.append

    def recc_von_neumann(arr, curr_dim=0, remaining_distance=distance, isCenter=True):
        #the breaking statement of the recursion
        if curr_dim == dimensions:
            if not isCenter:
                app(arr)
            return

        dimensions_coordinate = coordinates[curr_dim]
        if not (0 <= dimensions_coordinate < len(arr)):
            return 

        dimesion_span = range(dimensions_coordinate - remaining_distance, 
                              dimensions_coordinate + remaining_distance + 1)
        for c in dimesion_span:
            if 0 <= c < len(arr):
                recc_von_neumann(arr[c], 
                                 curr_dim + 1, 
                                 remaining_distance - abs(dimensions_coordinate - c), 
                                 isCenter and dimensions_coordinate == c)
        return
    
    recc_von_neumann(matrix)
    return neigh
```

```python
def zero_neighbors(arr,x,y):
    N = neighbors(arr,(x,y))
    return sum(N)
```

```python
def connected(arr):
    C = True
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i][j]==1 and zero_neighbors(arr,i,j)==0:
                C = False
    return C
```

```python
E_dict = {}
```

```python
from collections import Counter
from scipy.stats import entropy

for key in generations_trials.keys():
    trials = generations_trials[key]

    E = {trial:[] for trial in trials.keys()}

    for trial in trials.keys():
        for gen in trials[trial].keys():
            M = trials[trial][gen]
            Z = []
            for i in range(len(M)):
                for j in range(len(M)):
                    delta = 0
                    N=neighbors(M,(i,j))
                    for n in N:
                        if M[i][j]==n:
                            delta-=1/len(N)
                    Z+=[delta]
            C = list(dict(Counter(Z)).values())
            H = entropy(C,base = sum(C))
            E[trial] += [H]
    E_dict[key]=E
```

```python
E_list_dict = {}

for key in E_dict.keys():
    E_list = [list(E_dict[key][keykey]) for keykey in E_dict[key].keys()]
    E_list_dict[key]=[np.mean(t) for t in list(zip(*E_list))]
```

```python
AVG = [np.mean(x) for x in list(zip(*[list(E_list_dict[key]) for key in E_list_dict.keys()]))]
```
