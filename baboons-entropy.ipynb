```python
def neighbors(matrix, coordinates, distance=1):
    dimensions = len(coordinates)
    neigh = []
    app = neigh.append

    def recc_von_neumann(arr, curr_dim=0, remaining_distance=distance, isCenter=True):
        #the breaking statement of the recursion
        if curr_dim == dimensions:
            if not isCenter:
                app(arr)
            return

        dimensions_coordinate = coordinates[curr_dim]
        if not (0 <= dimensions_coordinate < len(arr)):
            return 

        dimesion_span = range(dimensions_coordinate - remaining_distance, 
                              dimensions_coordinate + remaining_distance + 1)
        for c in dimesion_span:
            if 0 <= c < len(arr):
                recc_von_neumann(arr[c], 
                                 curr_dim + 1, 
                                 remaining_distance - abs(dimensions_coordinate - c), 
                                 isCenter and dimensions_coordinate == c)
        return
    
    recc_von_neumann(matrix)
    return neigh
```

```python
def zero_neighbors(arr,x,y):
    N = neighbors(arr,(x,y))
    return sum(N)
```

```python
def connected(arr):
    C = True
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i][j]==1 and zero_neighbors(arr,i,j)==0:
                C = False
    return C
```

```python
from collections import Counter
from scipy.stats import entropy

for key in generations_trials.keys():
    trials = generations_trials[key]

    E = {trial:[] for trial in trials.keys()}

    for trial in trials.keys():
        for gen in trials[trial].keys():
            M = trials[trial][gen]
            Z = []
            for i in range(len(M)):
                for j in range(len(M)):
                    delta = 0
                    N=neighbors(M,(i,j))
                    #for s in range(3):
                    #    for t in range(3):
                    #        if N[1][1]==N[s][t]:
                    #            delta+=1/8
                    for n in N:
                        if M[i][j]==n:
                            delta-=1/len(N)
                    Z+=[delta]
            C = list(dict(Counter(Z)).values())
            H = entropy(C,base = sum(C))
            E[trial] += [H]
    E_dict[key]=E
```

```python
from itertools import product

def count_squares(A):
    width, height = len(A[0]), len(A)
    indices = product(range(width - 1), range(height - 1))
    is_square = lambda x, y: all(A[a][b] == 1 for a, b in product((x, x + 1), (y, y + 1)))
    return sum(1 for x, y in indices if is_square(x, y))
```

```python
N = 10

Z = [0]*(N**2)
```

```python
## condiciÃ³n inicial
import random
import numpy as np

list_matrices = []

for i in range(100):
    indexes = random.sample(range(len(Z)),N)
    I = []

    for i in range(len(Z)):
        if i in indexes:
            I+=[1]
        else:
            I+=[0]
        
    M = np.array(I).reshape((N,N))
    list_matrices+=[M]
```

```python
## evoluciÃ³n por energÃ­a!

def energy(array):
    Z = np.zeros((len(array),len(array)))
    for i in range(len(array)):
        for j in range(len(array)):
            delta = 0
            N=neighbors(array,(i,j))
            for n in N:
                if array[i][j]==n:
                    delta+=1/len(N)
            Z[i][j]=delta
    return -sum(sum(Z))/(len(array)*len(array))
```

```python
energy(M)
```

```python
def evol(M,T):

    energy_evol = []
    for t in range(T):
        energy_evol+=[energy(M)]
        Mt = M.copy()
        index_zero = []
        index_one = []
        for i in range(len(Mt)):
            for j in range(len(Mt)):
                if Mt[i][j]==0:
                    index_zero+=[(i,j)]
                else:
                    index_one+=[(i,j)]
        selected_zero = random.choice(index_zero)
        selected_one = random.choice(index_one)
        Mt[selected_zero[0]][selected_zero[1]]=1
        Mt[selected_one[0]][selected_one[1]]=0
        if energy(M)>energy(Mt):
            M = Mt.copy()
            
    return energy_evol
```
