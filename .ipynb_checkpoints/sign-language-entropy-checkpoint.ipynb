## 2. n-gramas

```python
STRINGS = {}

for key in datos.keys():
    L = datos[key]
    L = ['#'+s+'$' for s in L]
    STRINGS[key] = L
```

```python
STRINGS_rows = []

for item in datos_rows:
    L = item
    L = ['#'+s+'$' for s in L]
    STRINGS_rows += [L]
```

```python
STRINGS[0][:5]
```

```python
STRINGS_rows[0][:5]
```

```python
def find_ngrams(input_list, n):
    return [''.join(tup) for tup in list(zip(*[input_list[i:] for i in range(n)]))]
```

```python
find_ngrams('#kalu$',3)
```

```python
NGRAMAS = {}

for key in STRINGS.keys():
    ngramas = []
    for w in STRINGS[key]:
        ngramas+=find_ngrams(w,3)
    NGRAMAS[key] = ngramas
```

```python
NGRAMAS2 = {}

for key in STRINGS.keys():
    ngramas = []
    for w in STRINGS[key]:
        ngramas+=find_ngrams(w,2)
    NGRAMAS2[key] = ngramas
```

```python
NGRAMAS_rows = []

for item in STRINGS_rows:
    ngramas = []
    for w in item:
        ngramas+=find_ngrams(w,3)
    NGRAMAS_rows += [ngramas]
```

## 3. redes

```python
import networkx as nx

## función que construye redes
def GRAPH(D):
    G=nx.Graph()
    pairs = D
    for pair in pairs:
        for P in list(zip(pair,pair[1:])):
            
            if G.has_edge(P[0],P[1])==False:
                G.add_edge(P[0],P[1],weight=1)
            else:
                x=G[P[0]][P[1]]['weight']
                G[P[0]][P[1]]['weight']=x+1
        
    return G
```

```python
NGRAMAS[1][:5]
```

```python
grafos = {}

for key in NGRAMAS.keys():
    grafos[key] = GRAPH(NGRAMAS[key])
```

```python
from networkx.algorithms import bipartite

bipartivity = {}

for key in grafos.keys():
    G = grafos[key]
    bipartivity[key] = bipartite.spectral_bipartivity(G)
```

```python
betweenness = {}

for key in grafos.keys():
    G = grafos[key]
    betweenness[key] = np.mean(list(dict(nx.betweenness_centrality(G)).values()))
```
